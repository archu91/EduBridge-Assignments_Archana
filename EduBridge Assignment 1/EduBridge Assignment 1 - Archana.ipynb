 ## Life Expectancy Dataset

```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

1. Drop the irrelevant columns (drop Country column here).

```python
df.drop('Country', axis=1, inplace=True)
df.head()
```

2. Rename the mentioned columns ("Life expectancy": "Life_Expectancy", "Adult Mortality": "Adult_Mortality", "percentage expenditure": "Percentage_Exp").

```python
df.rename(columns={"Life expectancy": "Life_Expectancy", "Adult Mortality": "Adult_Mortality",
                      "percentage expenditure": "Percentage_Exp"}, inplace=True)
df.head(3)
```

3. Drop the rows in the data-set having duplicates.

```python
df.drop_duplicates(inplace=True)
df.head(3)
```

4. Drop the missing values in the data-set.

```python
df.dropna(inplace=True)
df.head(3)
```

Save your output file as eval.csv

```python
df.to_csv("eval.csv", index=False)
```

## Iris Dataset

Import the dataset iris.csv

```python
df = pd.read_csv("iris.csv", index_col=0)
df.head()
```

2. Create a new variable and name it Sepal.Area
3. Sepal.Area would contain multiplication of Sepal.Length and Sepal.Width

```python
df['Sepal.Area'] = df['Sepal.Length'] * df['Sepal.Width']
```

4. Subset the data by selecting Sepal.Length, Sepal.Width, Sepal.Area and Species

```python
df = df.iloc[:,[0,1,2,3,5,4]]
df.head()
```

5. Filter the data for Sepal.Area Values greater than 22.

```python
df = df[df['Sepal.Area']>22]
df
```

```python
df.to_csv('iris_new.csv', index=False)
```
